# Treating null values

There are both categorical and numerical columns which have missing values. so we will fill them by using appropriate methods.

The features Gender,Married,Dependents,Self-Employed,Loan_Amount_Term and Credit_History are categorical so will fill the missing values using their mode values.

```python
# Let's check the mode of the categorical columns to fill the null values
print('The mode of Gender is:', df['Gender'].mode())
print('The mode of Married is:', df['Married'].mode())
print('The mode of Dependents is:', df['Dependents'].mode())
print('The mode of Self_Employed is:', df['Self_Employed'].mode())
print('The mode of Loan_Amount_Term is:', df['Loan_Amount_Term'].mode())
print('The mode of Credit_History is:', df['Credit_History'].mode())
```

These are the modes of the categorical columns which contains null values.These are the values which are highly repeated in the columns. The missing values will be replaced by their respective mode values.

```python
# Filling the missing values of categorical columns

df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Married'] = df['Married'].fillna(df['Married'].mode()[0])
df['Dependents'] = df['Dependents'].fillna(df['Dependents'].mode()[0])
df['Self_Employed'] = df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mode()[0])
df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].mode()[0])
```

So the missing values in the categorical columns are filled using mode method.

```python

# Let's check the null values now
df.isnull().sum()
```

Lets fill the null values in numerical column i.e. Loan_Amount.

Loan_Amount has continuous data and has outliers and skewness too, so will fill the null values using median.

```python
# Let's checkk the median values of the numerical columns
print('The median of LoanAmount is:', df['LoanAmount'].median())
```

```python
# Filling null value in LoanAmount by its median value
df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].median())
```

# Zscore method

```python
# Numerical columns containing outliers
features = df[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term']]
from scipy.stats import zscore
z = np.abs(zscore(features))
z
```

```python
# Creating new dataframe
new_df = df[(z<3).all(axis = 1)]
new_df
```

```python
# Checking the dimensions of both dataframes
print(df.shape)
print(new_df.shape)
```

```python
# Checking the dataloss after removing the outliers
loss = (614-577)/614*100
loss
```

The data loss using zscore is 6.02%.

# IQR(InterQuartileRange)

```python
# 1st quartile
Q1 = features.quantile(0.25)

# 3rd quantile
Q3 = features.quantile(0.75)

#IQR
IQR = Q3 - Q1
df1 = df[~((df<(Q1 - 1.5 * IQR))|(df>(Q3 + 1.5 * IQR))).any(axis = 1)]
```

```python
df1
```

```python
df1.shape
```

```python
# Checking data loss percent after removing outliers using IQR
loss = (614-459)/614*100
loss
```

Using IQR method the data loss is 25%. So we will consider Zscore method.

# Modelling

# Finding the best random state

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
maxAccu = 0
maxRS = 0

for i in range(1,200):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state = i)
    RFC = RandomForestClassifier()
    RFC.fit(x_train, y_train)
    pred = RFC.predict(x_test)
    acc = accuracy_score(y_test, pred)
    if acc>maxAccu:
        maxAccu = acc
        maxRS = i
        
print('Best accuracy is', maxAccu, 'at Random_state', maxRS)
```

# Creating train_test_split

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state = maxRS)
```
